```python
# connecting to driver with data, after logging in using OBDC data source admin 
cnxn = pyodbc.connect(dsn= 'DBPROD')
```

```python
#printing table names available 
c = cnxn.cursor();

for row in c.tables():
    print (row.table_name)
```

```python
# writing a query to select information needed/wanted
query = '''
    Select 
    TOP 100
    ARACID,
    ARENID,
    ARACCLACCT,
   ZZACCLACCTCUST,
   ARACCURBAL,
   ARACNBTOTINI,
   ARRELSTATUSID AS STATUS, 
   ARRELPHASE AS PHASE, 
   FSDBPSELLERCONTID, 
   ZZOTHDBCAPTUREID, 
   ZZACPMIDR, 
   ZZACPMICR, 
   ARACSRVDTE, 
   ARACSTATUTEEXPDTE, 
   ARACLSTDTE, 
   ARRELSTATDTE, 
   ARACCANCDTE, 
   ARACCANCID, 
   ARACCANCBAKAMT, 
   ARACTYP, 
   ARACCOLTYP, 
   ARACLPYAMT, 
   ARACLPYDTE, 
   FSDBPSBREASONID, 
   ALACCLCRNAME,
   CURR_PLHIST.FSACPHVEID AS CURR_AGY,
   PREV_PLHIST.FSACPHVEID AS PREV_AGY 
   
   from          
   ARACCOUNT
   JOIN ARRELATIONSHIP ON ARACRPRELID = ARRELID
   JOIN ARENTITY ON ARACRPID = ARENID
   JOIN FSDBPURCHASE ON FSDBPID = FSACDBPURCHASEID
   LEFT JOIN ZZOTHDB ON ZZOTHDBACID = ARACID
   LEFT JOIN FSACPLHIST AS PREV_PLHIST ON FSACPREVPLID = PREV_PLHIST.FSACPHID
   LEFT JOIN FSACPLHIST AS CURR_PLHIST ON FSACCURRPLID = CURR_PLHIST.FSACPHID
   
   Where
   FSDBPSELLERCONTID = 773
   
   '''
```

```python
df = pd.read_sql(query, cnxn)
df.head()
```
